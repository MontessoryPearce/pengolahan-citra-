```

import skimage.io
import matplotlib.pyplot as plt
import numpy as np
import cv2

im2=skimage.io.imread('einstein.jpg')
im2=skimage.color.rgb2gray(im2)

def corr (im2,mask):
    row,col=im2[0].shape
    m,n=mask.shape
    new=np.zeros((row+m-1,col+n-1))
    n=n//2
    m=m//2
    filtered_im2=np.zeros(im2.shape)
    new[m:new.shape[0]-m,n:new.shape[1]-n]=im2[0]
    for i in range(n,new.shape[1]-n):
        for j in range (n,new.shape[1]-n):
            temp=new[i-m:i+m+1,j-m:j+m-1]
            result=temp*mask
            filtered_im2[i-m,j-n]=result
    return filtered_im2
np.set_printoptions(threshold=np.inf)    
print(im2)
```

```
import cv2
def GaussianBlur (m,n,nds):
    gaussian=np.zeros((m,n))
    m=m//2
    n=n//2
    nds=30.4
    for x in range(-m,m+1):
        for y in range (-n,n+1):
            x1=nds*(2*np.pi)**2
            x2=np.exp(-(x**2+y**2)/2(2*nds**2))
            gaussian[x+m,y+n]=(1/x1)*x2
    return gaussian
dst = cv2.GaussianBlur(im2,(5,5),0)

plt.subplot(121),plt.imshow(im2,'gray'),plt.title('original')

```
